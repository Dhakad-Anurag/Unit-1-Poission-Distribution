```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
placements_per_day = np.random.poisson(lam=4,size=4200)
```

```python
print(placements_per_day)
```

```python
np.sum(placements_per_day)
```

```python
x_axis = np.unique(placements_per_day)
```

```python
x_axis
```

```python
y_axis = list()

for x in x_axis:
    frequency = np.count_nonzero(placements_per_day== x)
    relative_frequency = frequency/placements_per_day.shape[0]
    
    y_axis.append(relative_frequency)
```

```python
y_axis
```

```python
plt.xlabel("Number of placements per day")
plt.ylabel("P(Placements/day)")
plt.bar(x_axis,y_axis)
```
